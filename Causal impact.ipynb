```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import random 
from causal_impact import CausalImpact
```

```python
data = pd.DataFrame(
     {'y': range(32),
      'x1': random.random() + random.random(),
      'x2': random.random()}
)
data
```

```python
ci = CausalImpact(data, 20, n_seasons=7)
ci.run(max_iter=1000)
ci.plot()
```
